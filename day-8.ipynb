```python
#day 8
# class - a blueprint for creating objects
class Bank():
    def __init__(self, balance=0):
        self.balance = balance
    def deposit(self,amount):
        if amount <= 0:
            print("Invalid")
        else:    
            self.balance += amount
            print(f"The amount is {self.balance}")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient amount")
        else:    
            self.balance -= amount
            print(f"The remaining amount is {self.balance}")

acc1 = Bank(10000)
acc1.deposit(1000)
acc1.withdraw(5000)
```

```python
class Rectangle:
    def __init__(self, height, width):
        self.height = height
        self.width = width
    def area(self):
        return self.height * self.width
    def perimeter(self):
        return 2 * (self.height + self.width)

r1 = Rectangle(2,4)
print("Area:", r1.area())
print("Perimeter:", r1.perimeter())
```

```python
#encadsulation
#1st we have current temperture and target temperture which is in private
# we use setter and getter function to change the target temperture 
# if the temp is in the limit then change the tar temp  
#
# 
#  #

class Thermostat:
    def __init__(self,cur,tar):
        self.__current_temp = cur
        self.__target_temp = tar

    def set(self, temp):
        if 16 <= temp <= 30:
            self.__target_temp = temp
            print(f"Target temperature set to {temp}°C")
        else:
            print("Invalid temperature range.")

    def get(self):
        print(f"Current: {self.__current_temp}°C, Target: {self.__target_temp}°C")

thermo = Thermostat(25,22)
thermo.set(24)
thermo.get()
```

```python
#inheritance
class Staff:
    def __init__(self, name, ID,dept):
        self.name = name
        self.ID = ID
        self.dept=dept

    def show_details(self):
        print(f"Name: {self.name}, ID: {self.ID},department: {self.dept}")

class Doctor(Staff):
    def diagnose(self):
        print(f"Dr. {self.name} is diagnosing a patient.")

class Nurse(Staff):
    def assist(self):
        print(f"Nurse {self.name} is assisting in surgery.")


d = Doctor("vishal", 101,"diagnose")
d.show_details()
d.diagnose()

n = Nurse("muthu", 102,"assist")
n.show_details()
n.assist()
```

```python
#polymorphism
class notifications:
    def send(self):
        pass
    
class Email(notifications):
    def send(self):
        print("Sending Email...")

class SMS(notifications):
    def send(self):
        print("Sending SMS...")

class Push(notifications):
    def send(self):
        print("Sending Push Notification...")

n1 = [Email(), SMS(), Push()]
for n in n1:
    n.send()
```
