```python
class atm():
    def __init__(s,pin):
    
        s.__pin=pin
    def get(s,pin1):
        if (pin1 >=1000 and pin1 <= 9999):#or (len(str(pin1)))==4:
            s.__pin=pin1
        else:
            print("Invaild pin")    
    def set(s):
        print(f"the pin is: {s.__pin}")
a=atm(1234)
a.set()
a.get(43221)
a.set()
```

```python
class person():
    def __init__(s,name,age):#constructor
        s.name=name #public
        s._age=age #protected
    def dis(s):
        print(f"the names {s.name} and the age is {s._age}")
    def set(s,new_age):
        if new_age>0:
            s._age=new_age 
    def get(s):
        return s._age  
p=person("godson",19)
p.dis() 
p.set(20) 
p.get()
p.dis()
```

```python
class animal():
    def sound(self):
        print("All animals make sound")
class dog(animal):
    def sound(self):
        print("Dog is barking")
class cat(animal):
    def sound(self):
        print("Cat is meow")

ani1 = [animal(),cat(),dog()]

for i in ani1:
    i.sound()
```

```python
#abstraction 
class juicemaker():
    def maker(s):
        s.__wash()
        s.__cutting()
        s.__blend()
        s.__add_suger()
        s.__serve()
    def __wash(s):
        print("wash the fruits")
    def __cutting(s):
        print("cut the fruits")
    def __blend(s):
        print("blend the fruits")
    def __add_suger(s):
        print("add suger to  the fruits juice")  
    def __serve(s):
        print("serve the fruits juice")   
jm=juicemaker()
jm.maker() 
  
```

```python
class Thermostat:
    def __init__(self,cur,tar):
        self.__current_temp = cur
        self.__target_temp = tar

    def set(self, temp):
        if 16 <= temp <= 30:
            self.__target_temp = temp
            print(f"Target temperature set to {temp}°C")
        else:
            print("Invalid temperature range.")

    def get(self):
        print(f"Current: {self.__current_temp}°C, Target: {self.__target_temp}°C")

thermo = Thermostat(25,22)
thermo.set(24)
thermo.get()
```

```python
#abstraction
class TravelBooking:
    def search(self, source, destination):
        print(f"Searching flights from {source} to {destination}...")

    def book(self, flight_id):
        print(f"Booking flight {flight_id}... Ticket booked!")

b = TravelBooking()
b.search("Chennai", "Delhi")
b.book("AI202")
```

```python
#inheritance
class Staff:
    def __init__(self, name, ID,dept):
        self.name = name
        self.ID = ID
        self.dept=dept

    def show_details(self):
        print(f"Name: {self.name}, ID: {self.ID},department: {self.dept}")

class Doctor(Staff):
    def diagnose(self):
        print(f"Dr. {self.name} is diagnosing a patient.")

class Nurse(Staff):
    def assist(self):
        print(f"Nurse {self.name} is assisting in surgery.")


d = Doctor("xx", 101,"diagnose")
d.show_details()
d.diagnose()

n = Nurse("yy", 102,"assist")
n.show_details()
n.assist()
```

```python
#polymorphism
class notifications:
    def send(self):
        pass
    
class Email(notifications):
    def send(self):
        print("Sending Email...")

class SMS(notifications):
    def send(self):
        print("Sending SMS...")

class Push(notifications):
    def send(self):
        print("Sending Push Notification...")

n1 = [Email(), SMS(), Push()]
for n in n1:
    n.send()
```
