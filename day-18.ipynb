```python
def pali(s):
    c=0
    n=len(s)
    for i in range(n):
        for j in range(i+1,n+1):
            if s[i:j]==s[i:j][::-1]:
                c+=1
    return c    

pali("abba")
```

```python
class WordBuilder:
    def __init__(self, s):
        self.s = s

    def build(self):
        return ''.join(self.s)


wb = WordBuilder(["com", "pu", "ter"])
print(wb.build())  
```

```python
class SpyChecker:
    def __init__(self):
        self.keywords = ["spy", "code", "agent"]

    def code(self, message):
        for word in self.keywords:
            if word in message:  
                return True
        return False
sc = SpyChecker()
print(sc.code("This message contains a secret agent"))  # Output: True
```

```python
class VowelCounter:
    def count(self, text):
        vowels = "aeiouAEIOU"
        c=0
        for i in text:
            if i in vowels:
                c+=1
        return c    
vc = VowelCounter()
print(vc.count("Hello, this is a clear voice script")) 
```
